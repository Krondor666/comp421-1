```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

n = 3000
```

A quick look at the softmax in 3D. A simplex!

```python
def softmax(z):
    return np.exp(z)/np.sum(np.exp(z),axis = 1).reshape((n,1))
```

```python
phi = np.random.standard_normal((n,3)) #initialise values from a gaussian dist
x = softmax(phi)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(16,16))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2],alpha = 0.1)
ax.view_init(elev=45.,azim=45)
```

```python
#Lets make a pretty plot of the entropy of our distribution for different random points.
def entropy(z):
    return -np.sum(z*np.log(z),axis =1)
ent = entropy(x)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(16,16))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2],alpha = 0.1,c = ent)
ax.view_init(elev=45.,azim=45)
```
