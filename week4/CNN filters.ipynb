```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import signal

from fuel.datasets import CIFAR10
cifar = CIFAR10(('train',))
state = cifar.open()
N = 5
req = [n for n in range(N)] 
im,labels = cifar.get_data(state=state, request=req)
reshape_ims = im.reshape((len(req),3,32,32))
```

```python
def plotter(W):
    #for a bunch of different images
    for i in range(N):
        plt.figure(figsize=(15,15))
        plt.subplot(1,2,1)
        plt.imshow(np.transpose(im[i],[1,2,0]))

        z = signal.convolve2d(reshape_ims[i,0,:,:],W,mode='same') #convolve with some filter
        plt.subplot(1,2,2)
        plt.imshow(z,cmap='Greys')
```

```python
# an edge detector
W = np.zeros((3,3))
W[1,0] = 1
W[0,1] = 1
W[2,1] = 1
W[1,2] = 1
W[1,1] = -4
```

```python
plotter(W)
```

```python
# an gradient detector
W = np.zeros((3,3))
W[0,0] = 1
W[0,1] = 1
W[0,2] = 1
W[2,0] = -1
W[2,1] = -1
W[2,2] = -1
```

```python
plotter(W)
```

```python
# a blurrer
W = np.ones((3,3))
```

```python
plotter(W)
```
